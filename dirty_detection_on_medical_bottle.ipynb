```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def enhance_color_and_contrast(image):
    # Apply color and contrast enhancements (you can adjust these as needed)
    enhanced_image = cv2.convertScaleAbs(image, alpha=1.5, beta=25)

    return enhanced_image

def edge_detection(image):
    # Convert the enhanced image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Canny edge detection (you can choose other edge detection methods)
    edges = cv2.Canny(gray, 50, 150, apertureSize=3)

    return edges

def main(image_path):
    # Read the image
    img = cv2.imread(image_path)

    # Check if the image was read successfully
    if img is None:
        print(f"Error: Unable to read the image at path {image_path}")
        return

    # Enhance color and contrast
    enhanced_img = enhance_color_and_contrast(img)

    # Apply edge detection
    edges = edge_detection(enhanced_img)

    # Display the original image, enhanced image, and edges
    cv2_imshow(img)
    cv2_imshow(enhanced_img)
    cv2_imshow(edges)

if __name__ == "__main__":
    image_path = "/content/botol.jpg"  # Replace with your image path
    main(image_path)
```
